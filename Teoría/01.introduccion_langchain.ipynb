## **LangChain** - _Introducción y Entorno con Gemini_

### 01. Historia de **LangChain**
**LangChain** fue creado por **Harrison Chase** hacia finales de octubre de 2022 como un proyecto de código abierto. Con el lanzamiento masivo de **ChatGPT** y la explosión de los _Large Language Models_ (**LLMs**) a principios de 2023, el proyecto experimentó un crecimiento sin precedentes, convirtiéndose rápidamente en la herramienta estándar de la industria para desarrollar aplicaciones basadas en IA.

Hoy en día, **LangChain** es una empresa consolidada que ofrece un ecosistema completo que incluye **LangChain** (orquestación), **LangSmith** (depuración y monitorización) y **LangGraph** (creación de agentes complejos basados en grafos).

### 02. ¿Qué es **LangChain** y por qué lo necesitamos?
**LangChain** es un **framework de orquestación** diseñado para simplificar la creación de aplicaciones que combinan **LLMs** con otras fuentes de cómputo o datos. 

Para entender su valor, debemos analizar la diferencia entre interactuar con un **LLM** de forma directa frente a usar una capa de abstracción:

* **Llamada Directa (SDK Propietario o API Rest):** Si programas directamente con el _Software Development Kit_ (**SDK**) de **Google GenAI** (u otros), tu código queda completamente acoplado a la estructura de datos, nombres de parámetros y métodos específicos de **Google**.
Si decides cambiar tu modelo a uno de **OpenAI**, **Anthropic** o un modelo local (como **Ollama**), tendrás que reescribir prácticamente toda la lógica de tu aplicación.

* **Capa de Abstracción (LangChain):** **LangChain** define una interfaz unificada. Para cambiar de proveedor de IA, solo necesitas cambiar la clase que instancia el modelo (por ejemplo, pasar de `ChatGoogleGenerativeAI` a `ChatOpenAI`). El resto de tu aplicación (prompts, cadenas, bases de datos vectoriales) permanece exactamente igual.

El nombre **LangChain** es un acrónimo de **Language** (Lenguaje) y **Chain** (Cadena), lo que describe perfectamente su función principal: enlazar (o encadenar) componentes modulares para procesar y actuar sobre el lenguaje.

### 03. Aplicaciones Prácticas

* **_Retrieval-Augmented Generation_ (RAG):** Conectar **LLMs** a manuales, PDFs o bases de datos de una empresa para responder preguntas con información privada y actualizada.

* **Agentes Autónomos:** Sistemas capaces de decidir por sí mismos qué herramientas usar (buscar en Google, ejecutar código Python, consultar una base de datos) para cumplir un objetivo.

* **Chatbots Avanzados:** Sistemas conversacionales con gestión de memoria a corto y largo plazo.

* **Extracción de Datos Estructurados:** Transformar texto libre (como correos o reseñas) en archivos **JSON** estructurados y tipados con total consistencia.

### 04. Ventajas y Desventajas

#### Ventajas:
* **Flexibilidad (Agnóstico al modelo):** Cambiar de proveedor de **LLM** requiere modificar una sola línea de código.
* **Ecosistema Gigante:** Integración nativa con cientos de herramientas (Bases de datos vectoriales, cargadores de archivos, herramientas de búsqueda).
* **_LangChain Expression Language_ (LCEL):** Permite construir flujos complejos mediante una sintaxis limpia y declarativa.

#### Desventajas:
* **Curva de aprendizaje:** La cantidad de conceptos y la abstracción pueden resultar confusas al principio.
* **Actualizaciones constantes:** Al ser una tecnología que evoluciona rápido, es común encontrar funciones obsoletas entre versiones.

Nunca escribas tus **API Keys** directamente en el código.

````python
import os
from pprint import pprint
from dotenv import load_dotenv

# Cargamos las variables de entorno del archivo .env al sistema operativo
load_dotenv()

# Verificamos que la variable se ha cargado
GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")

print(f"API Key: {GOOGLE_API_KEY[:7]}...")
````

### 07. Verificación de la Conexión

Vamos a instanciar nuestro modelo de chat utilizando la clase `ChatGoogleGenerativeAI`.

Usaremos el modelo `gemini-3.1-flash-lite` con el parámetro `temperature=0` que nos asegura respuestas deterministas.

````python
from langchain_google_genai import ChatGoogleGenerativeAI

MODEL = "gemini-3.1-flash-lite"
TEMPERATURE = 0

# Inicializamos el cliente del modelo
llm = ChatGoogleGenerativeAI(model = MODEL, temperature = TEMPERATURE)

# Realizamos una llamada de prueba utilizando el método estándar .invoke()
respuesta = llm.invoke(input = "Hola Gemini, ¿estás listo para el curso de LangChain?")

# Imprimimos el contenido de la respuesta
pprint(respuesta.content)
````

````python
print(respuesta.content[0]["text"])
````

### 08. Configuración de Parámetros

La clase `ChatGoogleGenerativeAI` (o cualquier otro proveedor) cuenta con distintos parámetros para configurar el comportamiento de la respuesta del modelo:

- `model`: str
    - Define qué variante del modelo de Google vas a utilizar.

- `google_api_key`: str/Opcional
    - La clave secreta para autenticarte en **Google AI Studio**, si este parámetro está vacío **LangChain** buscará automáticamente la variable de entorno `GOOGLE_API_KEY`

- `temperature`: float: 0.0 a 2.0/Opcional
    - Controla la aleatoriedad y la "creatividad" de las respuestas a nivel probabilístico.
    - **Valores bajos (0.0 a 0.2)**: El modelo seleccionará siempre las palabras más probables. La respuesta será lógica, repetitiva y predecible. Es obligatorio usar 0.0 para extraer **JSON**, procesar datos estructurados o escribir código.
    - **Valores altos (0.7 a 1.2)**: El modelo se vuelve más variado y creativo, arriesgándose con combinaciones de palabras menos comunes. Ideal para redactar correos, generar nombres de marcas o escribir ficción.

- `top_p`: float: 0.0 a 1.0/Opcional
    - Conocido como _Nucleus Sampling_. Funciona en conjunto con la temperatura. Filtra las palabras que el modelo considera para su respuesta según la probabilidad acumulada.
    - Si fijas `top_p=0.9`, el modelo solo evaluará el 90% de las palabras más probables y descartará el 10% restante, evitando respuestas completamente absurdas o inconexas incluso con temperaturas altas.

- `top_k`: int/Opcional
    - Limita el número estricto de palabras que el modelo analiza para generar el siguiente token.
    - Si fijas `top_k=40`, el modelo solo elegirá la siguiente palabra entre sus 40 mejores opciones, sin importar lo alta que sea la temperatura. Reduce la posibilidad de que el modelo comience a "alucinar" o desvariar.

- `max_output_tokens`: int/Opcional
    - Establece el límite máximo de tokens (aproximadamente palabras) que el modelo puede generar en una sola respuesta.
    - Es un salvaguardas económico indispensable para evitar que un bucle o una mala instrucción generen respuestas innecesariamente largas que agoten tu cuota de la API.

````python
# Configuración avanzada

MODEL = "gemini-3.1-flash-lite"
TEMPERATURE = 0.8
TOP_P = 0.95
TOP_K = 40
MAX_OUTPUT_TOKENS = 500

llm_params = {"model" : MODEL,
              "temperature" : TEMPERATURE,
              "top_p" : TOP_P,
              "top_k" : TOP_K,
              "max_output_tokens" : MAX_OUTPUT_TOKENS}

llm_avanzado = ChatGoogleGenerativeAI(**llm_params)

respuesta = llm_avanzado.invoke(input = "Hola Gemini, ¿estás listo para el curso de LangChain?")

print(respuesta.content[0]["text"])
````

### 09. Ecosistema de Proveedores en LangChain

Una de las grandes ventajas de LangChain es su arquitectura modular. Si en el futuro necesitas cambiar a otro modelo, solo necesitas instalar su paquete correspondiente y cambiar la clase de instanciación. El resto de tu código permanecerá igual.

|Proveedor       |Modelos Destacados                    |Clase en LangChain      |Comando de Instalación              |
|----------------|--------------------------------------|------------------------|------------------------------------|
|**Google**      |Gemini 3.1 Flash, Gemma 4 31B         |`ChatGoogleGenerativeAI`|`pip install langchain-google-genai`|
|**OpenAI**      |GPT-4o, GPT-4-turbo, GPT-3.5          |`ChatOpenAI`            |`pip install langchain-openai`      |
|**Anthropic**   |Claude 3.5 Sonnet, Opus, Haiku        |`ChatAnthropic`         |`pip install langchain-anthropic`   |
|**Mistral AI**  |Mistral Large, Mixtral, Mistral 7B    |`ChatMistralAI`         |`pip install langchain-mistralai`   |
|**Groq**        |Llama 3, Mixtral                      |`ChatGroq`              |`pip install langchain-groq`        |
|**Cohere**      |ommand R, Command R+                  |`ChatCohere`            |`pip install langchain-cohere`      |
|**AWS Bedrock** |Ecosistema AWS (Claude, Llama, Titan) |`ChatBedrock`           |`pip install langchain-aws`         |
|**Ollama**      |Modelos locales (Llama 3, Phi-3)      |`ChatOllama`            |`pip install langchain-ollama`      |
|**Hugging Face**|Modelos Open Source del Hub           |`ChatHuggingFace`       |`pip install langchain-huggingface` |

*Nota: Para usar cualquiera de estos proveedores, deberás agregar su respectiva API Key en tu archivo `.env` (ej. `OPENAI_API_KEY`, `ANTHROPIC_API_KEY`), excepto para Ollama, que se ejecuta localmente en tu propia máquina.*

````python
# Mismo código, pero con Groq

from langchain_groq import ChatGroq

load_dotenv()

MODEL = "openai/gpt-oss-120b"
TEMPERATURE = 0

# Inicializamos el cliente del modelo
llm = ChatGroq(model = MODEL, temperature = TEMPERATURE)

# Realizamos una llamada de prueba utilizando el método estándar .invoke()
respuesta = llm.invoke(input = "Hola Gemini, ¿estás listo para el curso de LangChain?")

# Imprimimos el contenido de la respuesta
pprint(respuesta.content)
````
