## **LangChain** - _Loaders_

El objetivo de un `Loader` es tomar datos crudos y convertirlos en objetos `Document` estandarizados, los cuales contienen:
1. `page_content`: El texto extraído.
2. `metadata`: Información sobre el origen del documento (ruta, página, fila, etc.), crucial para el filtrado en bases de datos vectoriales.

### 01. Plain Text (_.txt_)
El formato más sencillo. Ideal para logs, transcripciones crudas o notas. Utilizaremos `TextLoader`.

```python
from langchain_community.document_loaders import TextLoader

# Es una buena práctica siempre especificar el encoding para evitar errores de lectura
txt_loader = TextLoader(file_path = "../Data/ejemplo_documento.txt", encoding = "utf-8")
docs_txt = txt_loader.load()

print(f"Se cargaron {len(docs_txt)} documentos de texto.")
print(docs_txt[0])
```

### 02. Comma-Separeted Values (_.csv_)
Para datos tabulares, `CSVLoader` es excelente. Por defecto, carga cada fila del **csv** como un documento independiente, lo cual es perfecto para mantener el contexto aislado por registro.

```python
from langchain_community.document_loaders import CSVLoader

# Recomendación 2026: Usa "source_column" para que los metadatos apunten a un ID único
csv_loader = CSVLoader(file_path = "../Data/airbnb_madrid.csv",
                       source_column = "id",
                       encoding = "utf-8")
docs_csv = csv_loader.load()

print(f"Se cargaron {len(docs_csv)} filas del csv.")
print(f"Metadatos de la primera fila: {docs_csv[0].metadata}")
print(f"Datos de la primera fila: \n{docs_csv[0].page_content}")
```

```python
docs_csv[0].page_content
```

### 03. JavaScript Object Notation (_.json_)
Para estructuras anidadas, **LangChain** utiliza `JSONLoader`, el cual se apoya en la sintaxis de `jq` para parsear y extraer exactamente el nodo de información que necesitas.

```python
from langchain_community.document_loaders import JSONLoader

# jq_schema=".[]" iterará sobre la lista donde empiezan los datos del archivo .json
json_loader = JSONLoader(file_path = "../Data/tv_shows_100.json",
                         jq_schema = ".[]",
                         text_content = False)

docs_json = json_loader.load()

print(f"Se extrajeron {len(docs_json)} nodos del JSON.")
```

```python
docs_json[0].page_content
```

### 04. PDF (_.pdf_)
Los PDFs pueden ser complejos por su formato visual. `PyPDFLoader` es el estándar por su equilibrio entre velocidad y precisión. Una ventaja es que divide automáticamente el documento por páginas.

```python
from langchain_community.document_loaders import PyPDFLoader

pdf_loader = PyPDFLoader(file_path = "../Data/attention-is-all-you-need.pdf")

docs_pdf = pdf_loader.load()

print(f"Se cargaron {len(docs_pdf)} páginas del PDF.")
print(f"Metadatos (Página {docs_pdf[0].metadata["page"]}): {docs_pdf[0].metadata["source"]}")
```

```python
print(docs_pdf[1].page_content)
```

```python
print(docs_pdf[0].page_content)
```

### 05. HyperText Markup Language (_.html_)
Para extraer texto de páginas web descargadas sin traer todo el ruido de las etiquetas de estilo o scripts, usamos `BSHTMLLoader` (basado en `BeautifulSoup`).

```python
from langchain_community.document_loaders import BSHTMLLoader

html_loader = BSHTMLLoader(file_path = "../Data/ejemplo_html.html", open_encoding = "utf-8")
docs_html = html_loader.load()

print(f"Archivo HTML procesado.")
# El loader limpia las etiquetas HTML y extrae el título en los metadatos
print(f"Título extraído: {docs_html[0].metadata["title"]}")
```

```python
docs_html[0].metadata
```

```python
print(docs_html[0].page_content)
```

### 06. **DirectoryLoader**

`DirectoryLoader` busca dentro de una carpeta los archivos que le indiques y luego cargar cada archivo con algún **Loader** que vimos anteriormente.

```python
from langchain_community.document_loaders import DirectoryLoader

# Carga todos los archivos .txt dentro de la carpeta Data
dir_loader = DirectoryLoader(path = "../Data/coches",
                             glob = "**/*.json",
                             loader_cls = JSONLoader,
                             loader_kwargs = {"jq_schema": ".", "text_content": False},
                             show_progress = True)

docs_dir = dir_loader.load()

print(f"Se cargaron {len(docs_dir)} documentos desde el directorio.")
```

```python
docs_dir[0].metadata
```

```python
eval(docs_dir[0].page_content)
```
